```python
df = pd.DataFrame(
    {
        'ad': ['Egemen', 'Hakan', 'Ayşe', 'Faruk', 'Bülent', 'Levent','Jale'],
        'yaş': [23, 78, 22, 19, 45, 33, 20],
        'cinsiyet': ['erkek','erkek','kadın','erkek','erkek','erkek','kadın'],
        'şehir': ['Ankara','İstanbul','Bursa','Adana','İstanbul','Ankara','Ankara'],
        'poliçe_yıl': [1,3,2,4,5,6,3],
        'kaza_sayısı': [1,0,0,5,3,6,1]
    }
)
df
```

```python
#df.plot(kind='scatter', x='yaş', y='kaza_sayısı', color='red')
plt.scatter(df.yaş, df.kaza_sayısı, color='red')
```

```python
sayilar = df.groupby("şehir").sum()
sayilar
```

```python
%matplotlib inline
plt.pie(x=sayilar.kaza_sayısı, labels=sayilar.index)
plt.show()
```

```python
sayilar2 = df.groupby("cinsiyet").sum()
%matplotlib inline
plt.pie(x=sayilar2.kaza_sayısı, labels=sayilar2.index)
plt.show()
```

```python
%matplotlib inline
plt.bar(df["yaş"], df["kaza_sayısı"], label='poliçe yıl sayısı', color='g')
plt.bar(df["yaş"], df["poliçe_yıl"], label='kaza sayısı', color='r')
plt.legend()
plt.show()
```

```python
eth = pd.read_csv('ethereum.csv', usecols=["Date","Close"])
btc = pd.read_csv('bitcoin.csv', usecols=["Date","Close"])
bnb = pd.read_csv('BNB.csv', usecols=["Date","Close"])

eth.rename(columns = {'Close': "eth"}, inplace = True)
btc.rename(columns = {'Close': "btc"}, inplace = True)
bnb.rename(columns = {'Close': "bnb"}, inplace = True)

df = pd.merge(pd.merge(eth, btc, on='Date'),bnb, on='Date')
df["Date"] = pd.to_datetime(df["Date"])
df.dtypes
```

```python
plt.figure(figsize = (10,10))
plt.plot(df.Date, df.btc, 'r.-', label='BTC')
plt.plot(df.Date, df.bnb, 'g.-', label='BNB')
plt.plot(df.Date, df.eth, 'b.-', label='ETH')
plt.legend()
plt.show()
```
